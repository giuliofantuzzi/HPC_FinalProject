```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing csv files

```python
bcast_default_df= pd.read_csv('bcast_default.csv')
bcast_linear_df= pd.read_csv('bcast_linear.csv')
bcast_chain_df= pd.read_csv('bcast_chain.csv')
bcast_binarytree_df= pd.read_csv('bcast_binarytree.csv')
```

and merging them into a single dataframe

```python
bcast_df= pd.concat([bcast_default_df, bcast_linear_df, bcast_chain_df, bcast_binarytree_df], ignore_index=True)
```

# Avg Latency vs cores (map by cores)

## (1) All cores between 1 and 256

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Processes', y='AvgLatency(us)', hue='Algorithm', data=bcast_df)
plt.xlabel('N. of processes (#cores)')
plt.title('Broadcast Avg Latency vs #cores (map-by core and message size= 1 MPI CHAR)\n')
# add vertical lines
plt.axvline(x=16, color='grey', linestyle='--')
plt.axvline(x=32, color='grey', linestyle='--')
plt.axvline(x=64, color='grey', linestyle='--')
plt.axvline(x=128, color='grey', linestyle='--')

# Add text near the vertical lines
plt.text(16, plt.ylim()[0], '16', va='bottom', ha='left', color='grey')
plt.text(32, plt.ylim()[0], '32', va='bottom', ha='left', color='grey')
plt.text(64, plt.ylim()[0], '64', va='bottom', ha='left', color='grey')
plt.text(128, plt.ylim()[0], '128', va='bottom', ha='left', color='grey')
```

## (2) Powers of 2

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='Processes', y='AvgLatency(us)', hue='Algorithm', data=bcast_df[bcast_df['Processes'].isin([2,4,8,16, 32,64,128,256])])
plt.xlabel('N. of processes (#cores)')
plt.title('Broadcast Avg Latency vs #cores (map-by core and message size= 1 MPI CHAR)\n')
# add vertical lines
plt.axvline(x=16, color='grey', linestyle='--')
plt.axvline(x=32, color='grey', linestyle='--')
plt.axvline(x=64, color='grey', linestyle='--')
plt.axvline(x=128, color='grey', linestyle='--')

# Add text near the vertical lines
plt.text(16, plt.ylim()[0], '16', va='bottom', ha='left', color='grey')
plt.text(32, plt.ylim()[0], '32', va='bottom', ha='left', color='grey')
plt.text(64, plt.ylim()[0], '64', va='bottom', ha='left', color='grey')
plt.text(128, plt.ylim()[0], '128', va='bottom', ha='left', color='grey')
```
