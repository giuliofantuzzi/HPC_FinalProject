```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bcast_standard_df = pd.read_csv('../bcast_results/bcast_standard.csv')
bcast_chain_df = pd.read_csv('../bcast_results/bcast_chain.csv')
bcast_binarytree_df = pd.read_csv('../bcast_results/bcast_binarytree.csv')
```

```python
# merge the dataframes
bcast_df = pd.concat([bcast_standard_df, bcast_chain_df, bcast_binarytree_df])
bcast_df.columns
```

```python
bcast_df.shape
```

```python
# lineplot coloring by alg
plt.figure(figsize=(12, 8))
sns.lineplot(x='Processes', y='AvgLatency(us)', hue='Algorithm', data=bcast_df,
             palette=["#3498DB", "#F39C12", "#A93226"])
plt.title('Bcast Avg latency vs #cores (map-by core and message size=1 MPI CHAR)')
plt.xlabel('N. of processes (#cores)')
```

```python
# lineplot coloring by alg
plt.figure(figsize=(12, 8))
sns.lineplot(x='Processes', y='MinLatency', hue='Algorithm', data=bcast_df,
             palette=["#3498DB", "#F39C12", "#A93226"])
plt.title('Bcast Min latency vs #cores (map-by core and message size=1 MPI CHAR)')
plt.xlabel('N. of processes (#cores)')
```
