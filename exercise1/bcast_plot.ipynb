```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bcast_df= pd.read_csv('results/bcast_performance.csv')
```

```python
#lineplot of latency vs processes coloring by algorithm
sns.lineplot(x='Processes', y='Avg Latency(us)', hue='Algorithm', data=bcast_df,
             palette=['red', 'orange','blue'])
# change legend names
plt.xlabel('Number of Processes (#cores)')
plt.ylabel('Average Latency (us)')
plt.title('Broadcast Latency vs #cores')
plt.legend(title='Algorithm', loc='upper left')
#theme
sns.set_style('whitegrid')
```
