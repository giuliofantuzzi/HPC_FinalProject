```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
bcast_chain_df=pd.read_csv("bcast_chain.csv")
bcast_linear_df=pd.read_csv("bcast_linear.csv")
bcast_df= pd.concat([bcast_chain_df, bcast_linear_df], ignore_index=True)
```

```python
sns.lineplot(x='Processes', y='AvgLatency(us)', hue='Algorithm', data=bcast_chain_df)
plt.axvline(x=12, color='grey', linestyle='--')
plt.axvline(x=24, color='grey', linestyle='--')
```

**NON MI TORNANO I SALTI CAZZO!!!**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Processes', y='AvgLatency(us)', hue='Algorithm', data=bcast_df)
plt.xlim(2,48)

plt.xlabel('N. of processes (#cores)')
plt.title('Broadcast Avg Latency vs #cores (map-by core and message size= 1 MPI CHAR)\n')
plt.legend(loc='upper left',title="Algorithm")
# add vertical lines
plt.axvline(x=12, color='grey', linestyle='--')
plt.axvline(x=24, color='grey', linestyle='--')
# Add text near the vertical lines
plt.text(12, plt.ylim()[0], '12', va='bottom', ha='left', color='grey')
plt.text(24, plt.ylim()[0], '24', va='bottom', ha='left', color='grey')
```
