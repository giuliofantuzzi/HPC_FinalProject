```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
bcast_chain_df=pd.read_csv("bcast_chain.csv")
bcast_linear_df=pd.read_csv("bcast_linear.csv")
bcast_default_df=pd.read_csv("bcast_default.csv")
bcast_binarytree_df=pd.read_csv("bcast_binarytree.csv")

bcast_df= pd.concat([bcast_chain_df, bcast_linear_df,bcast_default_df,bcast_binarytree_df], ignore_index=True)
```

**NON MI TORNANO I SALTI CAZZO!!!**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Processes', y='AvgLatency(us)', hue='Algorithm', data=bcast_df[bcast_df['Processes']%2==0])
plt.xlim(2,48)

plt.xlabel('N. of processes (#cores)')
plt.title('Broadcast Avg Latency vs #cores (map-by core and message size= 1 MPI CHAR)\n')
plt.legend(loc='upper left',title="Algorithm")
# add vertical lines
plt.axvline(x=12, color='grey', linestyle='--')
plt.axvline(x=16, color='red', linestyle='--')
plt.axvline(x=24, color='grey', linestyle='--')
plt.axvline(x=32, color='red', linestyle='--')

# Add text near the vertical lines
plt.text(12, plt.ylim()[0], '12', va='bottom', ha='left', color='grey')
plt.text(16, plt.ylim()[0], '16', va='bottom', ha='left', color='red')
plt.text(24, plt.ylim()[0], '24', va='bottom', ha='left', color='grey')
plt.text(32, plt.ylim()[0], '32', va='bottom', ha='left', color='red')

sns.set_theme(style="white")
```
