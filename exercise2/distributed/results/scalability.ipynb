```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bcast_FlatTree_df= pd.read_csv('bcast_FlatTree.csv')
bcast_Chain_df= pd.read_csv('bcast_Chain.csv')
bcast_df= pd.concat([bcast_FlatTree_df, bcast_Chain_df], ignore_index=True)

```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Processes', y='AvgTime', hue='Algorithm', data=bcast_df)
# add line y=0.15*x

plt.xlabel('N. of processes (#cores)')
plt.title('Broadcast AvgTime vs #cores (1 MPI_INT)\n')
plt.legend(loc='upper left',title="Algorithm")
# add vertical lines
plt.axvline(x=16, color='grey', linestyle='--')
plt.axvline(x=32, color='grey', linestyle='--')
plt.axvline(x=64, color='grey', linestyle='--')
plt.axvline(x=128, color='grey', linestyle='--')

plt.text(16, plt.ylim()[0], '16', va='bottom', ha='left', color='grey')
plt.text(32, plt.ylim()[0], '32', va='bottom', ha='left', color='grey')
plt.text(64, plt.ylim()[0], '64', va='bottom', ha='left', color='grey')
plt.text(128, plt.ylim()[0], '128', va='bottom', ha='left', color='grey')
plt.text(256, plt.ylim()[0], '256', va='bottom', ha='left', color='grey')
```
